Write a function
def surround(word,n,symbol)
that returns the string word surrounded by the string symbol, repeated n times on both sides. 
For example,. python surround("wow",3,"S") return"SSSwowSSS"

```python
def surround(word,n,symbol):
    surrounding = symbol*n
    return surrounding +word +surrounding
result = surround("wow",3,"S")
print(result)
```

Write a function
def countVowels(String)
that returns a count of all vowels in the string string. Vowels are the letters a,e,i,o and u and their uppercase variants

```python
def countVowels(string):
    vowels="AEIOUaeiou"
    vowel_count=0
    for char in string:
        if char in vowels:
            vowel_count +=1
    return vowel_count
result = countVowels("Hello World")
print(result)
```

Write a function called print_fibonacci that takes a number as an argument and prints the Fibonacci sequence up to that number. The Fibonacci sequence is a series of numbers where each number is the sum of the 2 preceding ones. For example, if the number is 8, the function should print:0 1 1 2 3 5 8

```python
def print_fibonacci(number):
    a,b=0,1
    while a<= number:
        print(a,end=" ")
        a,b=b,a+b
print_fibonacci(8)
```

Write a recursive function called sum_digits that takes a positive integer as an argument
and returns the sum of its digits. For example, sum_digits(123) should return 1+2+3 =6

```python
def sum_digits(n):
    if n<10:
        return n
    else:
        return(n%10)+ sum_digits(n//10)
result = sum_digits(123)
print(result)
```

Use recursion to compute a^n, where n is a positive integer
Hint: if n is 1, then a^n=a. If n is even, then a^n =(a^n/2)^2

```python
def power(a,n):
    if n==1:
        return a
    elif n%2==0:
        half_power = power(a,n//2)
        return half_power * half_power
    else:
        return a*power(a,n-1)
result= power(2,5)
print(result)
```
